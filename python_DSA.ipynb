**Arrays:**
In Python, arrays are typically handled using lists or the array module. Here's an overview of both approaches:

1. Using Lists (most common):

Lists in Python are dynamic arrays, which means they can grow or shrink in size. Lists can contain elements of any type.

```python
import array

# Creating an array
arr = array.array('i', [1, 2, 3, 4, 5])  # 'i' indicates integers

# Accessing elements
print(arr[0])  # Outputs: 1

# Adding elements
arr.append(6)

# Removing elements
arr.remove(3)

# Slicing
print(arr[1:4])  # Outputs: array('i', [2, 4, 5])

# Iterating over an array
for item in arr:
    print(item)
```

**Stack:**
In Python, a stack is typically implemented using a list or the collections.deque module. A stack follows the Last In, First Out (LIFO) principle, meaning the last element added is the first one to be removed.

1. Using a List (most common method):

Python lists provide an easy way to implement a stack using append() to push items onto the stack and pop() to remove the top item from the stack.

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        else:
            return "Stack is empty"

    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        else:
            return "Stack is empty"

    def is_empty(self):
        return len(self.stack) == 0

# Using the stack
my_stack = Stack()
my_stack.push(10)
my_stack.push(20)
my_stack.push(30)

print("Top element:", my_stack.peek())  # Output: 30
print("Popped element:", my_stack.pop())  # Output: 30
print("Is stack empty?", my_stack.is_empty())  # Output: False
```
